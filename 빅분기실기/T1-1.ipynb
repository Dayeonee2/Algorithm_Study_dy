```python
import pandas as pd

df = pd.read_csv('dataset/titanic/train.csv')
```

```python
print(df.shape)
print(df.isnull().sum())
df.head()
```

```python
Q1 = df.Fare.quantile(0.25)
Q3 = df.Fare.quantile(0.75)
IQR = Q3 - Q1
Q1-1.5*IQR, Q3+1.5*IQR
```

```python
outdata=df[(df.Fare < Q1-1.5*IQR) | (df.Fare > Q3+1.5*IQR)]
outdata
```
